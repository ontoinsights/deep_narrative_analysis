# Extracting WordNet Synonyms, Hypernyms

```python
with open("data-noun.txt", "r") as wn:
    wn_text = wn.read()
wn_detail = wn_text.split("\n")

syn_noun_dict = dict()
domain_topics = set()
for syn_detail in wn_detail:
    if syn_detail.startswith("synOffset"):
        continue
    syn_split = syn_detail.split(" ")
    if len(syn_split) < 3:    # Blank line 
        continue
    syn = syn_split[0]
    number_words = int(syn_split[3])
    words = []
    for i in range(4, 4 + (number_words * 2), 2):
        words.append(f"{syn_split[i]}{syn_split[i+1]}")
    instance_hypernyms = []
    hypernyms = []
    if " @" in syn_detail:
        at_split = syn_detail.split(" @")
        for i in range(1, len(at_split)):
            hyper_split = at_split[i].split(" ")
            if hyper_split[2] == "n" and hyper_split[3] == "0000":
                if hyper_split[0].startswith("i "):
                    instance_hypernyms.append(f"{hyper_split[1]}")
                else:
                    hypernyms.append(f"{hyper_split[1]}")
    related_verbs = []
    if " + " in syn_detail:
        plus_split = syn_detail.split(" +")
        for i in range(1, len(plus_split)):
            rel_split = plus_split[i].split(" ")
            if rel_split[1] == 'v':
                related_verbs.append(f"{rel_split[0]}")
    gloss = syn_detail.split(" | ")[1].strip()
    topics = []
    if ";c " in syn_detail:
        topic_split = syn_detail.split(";c ")
        for i in range(1, len(topic_split)):
            c_split = topic_split[i].split(" ")[0]
            topics.append(c_split)
            domain_topics.add(c_split)
    syn_noun_dict[syn] = (words, hypernyms, instance_hypernyms, related_verbs, topics, gloss)

topics_text = []
for topic in domain_topics:
    topics_text.append(f"{syn_noun_dict[topic][0]} {topic}")
sorted_noun_topics = list(topics_text)
sorted_noun_topics.sort()
print(len(sorted_noun_topics))

i = 0
for key, value in syn_noun_dict.items():
    i += 1
    if i > 25:
        break
    print(key, value)
for i in range(0,25):
    print(sorted_noun_topics[i])
```

# Synset Hierarchy

```python
with open("data-verb.txt", "r") as wn:
    wn_text = wn.read()
wn_detail = wn_text.split("\n")

with open("verb-hierarchy.ttl", "w") as ttl:
    for syn_detail in wn_detail:
        if syn_detail.startswith("synOffset"):
            continue
        syn_split = syn_detail.split(" ")
        if len(syn_split) < 3:    # Blank line 
            continue
        syn = syn_split[0]
        number_words = int(syn_split[3])
        words = []
        more_words = []
        for i in range(4, 4 + (number_words * 2), 2):
            if syn_split[i+1] != '0':
                more_words.append(f"{syn_split[i]}")
            else:
                words.append(f"{syn_split[i]}")
        hypernyms = []
        if " @" in syn_detail:
            at_split = syn_detail.split(" @")
            for i in range(1, len(at_split)):
                hyper_split = at_split[i].split(" ")
                if hyper_split[2] == "v" and hyper_split[3] == "0000":
                    if hyper_split[0].startswith("i "):
                        continue
                    else:
                        hypernyms.append(f"{hyper_split[1]}")
        ttl.write(f"urn:{syn} a urn:Synset .\n")
        for hyper in hypernyms:
            ttl.write(f"urn:{syn} rdfs:subClassOf urn:{hyper} .\n")
        for word in words:
            ttl.write(f'urn:{syn} rdfs:label "{word}" .\n')
        for word in more_words:
            ttl.write(f'urn:{syn} skos:altLabel "{word}" .\n')

# Repeated for data-noun to create noun-hierarchy.ttl
```
