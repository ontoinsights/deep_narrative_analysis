```python
from lark import Lark, Tree

with open('lark.txt', 'r') as inFile:
    lark_details = inFile.read()
dsl = Lark(lark_details)

rule = ":Possession"
print('class_name', rule)
rule_parse = dsl.parse(rule)
tree = rule_parse.children[0].children[0]
print(tree)
class_name = tree.children[0].value
print(class_name)
print()

rule = "xcomp > :Attempt, xcomp attack"
print('xcomp', rule)
rule_parse = dsl.parse(rule)
tree = rule_parse.children[0].children[0]
print(tree)
class_tree, verb_tree = tree.children
print(class_tree.children[0].value)
xcomp, text = verb_tree.children
print(text.children[0].value)   
print()

rule = "'back' & dobj > :ReturnRecoveryAndRelease ; :has_topic dobj"
print('keyword', rule)
rule_parse = dsl.parse(rule)
tree = rule_parse.children[0].children[0]
print(tree)       
if len(tree.children) == 4:
    text_tree, objects_tree, class_tree, property_tree = tree.children
else:
    text_tree, objects_tree, class_tree = tree.children
    property_tree = None
looking_for = ''
for text in text_tree.children:
    looking_for += f'{text.value} '
print(looking_for[:-1])
if objects_tree:
    print(objects_tree.children[0].data)
print(class_tree.children[0].value)
if property_tree:
    print(property_tree)
    prop_names, prop_obj_type = property_tree.children
    prop_name = ':'
    for pname in prop_names.children:
        prop_name += f'{pname.value}_'
    print(prop_name[:-1])
    print(prop_obj_type.children[0].data)  
print()
    
rule = "prep_for (Agent pobj > :CaringForDependent ; :has_affected_agent pobj)"
print('preposition', rule)
rule_parse = dsl.parse(rule)
tree = rule_parse.children[0].children[0] 
print(tree)
prep_tree, complex_rule = tree.children
print(prep_tree.value)
print(complex_rule)
print()

rule = "Agent dobj > :CaringForDependents ; :has_affected_agent dobj"
# rule = "dobj > :Ignore, dobj verb"
print('object', rule)
rule_parse = dsl.parse(rule)
tree = rule_parse.children[0].children[0] 
print(tree)
entity_tree = tree.children[0]
obj1_tree = tree.children[1]
obj2_tree = tree.children[2]
class_tree = tree.children[3]
prop_or_verbs = []
if len(tree.children) > 4:
    for i in range(4, len(tree.children)):
        prop_or_verbs.append(tree.children[i]) 
if entity_tree:
    print(entity_tree.children[0].data)
print(obj1_tree.children[0].data)
if obj2_tree:
    print(obj2_tree.children[0].data)
print(class_tree.children[0].value)
for prop_or_verb in prop_or_verbs:
    if prop_or_verb.children[0].data == 'property_text':   
        prop_names, prop_obj_type = prop_or_verb.children[0].children
        prop_name = ':'
        for pname in prop_names.children:
            prop_name += f'{pname.value}_'
        print(prop_name[:-1])
        print(prop_obj_type.children[0].data)  # Was 1 children[]
    else:
        obj_type, text = prop_or_verb.children[0].children
        print(obj_type.data)   
        print(text.children[0].value)
print()
        
rule = "dobj & pobj > :Storage ; :has_topic dobj ; :has_instrument pobj"
print('object', rule)
rule_parse = dsl.parse(rule)
tree = rule_parse.children[0].children[0] 
print(tree)
entity_tree = tree.children[0]
obj1_tree = tree.children[1]
obj2_tree = tree.children[2]
class_tree = tree.children[3]
prop_or_verbs = []
if len(tree.children) > 4:
    for i in range(4, len(tree.children)):
        prop_or_verbs.append(tree.children[i]) 
if entity_tree:
    print(entity_tree.children[0].data)
print(obj1_tree.children[0].data)
if obj2_tree:
    print(obj2_tree.children[0].data)
print(class_tree.children[0].value)
for prop_or_verb in prop_or_verbs:
    if prop_or_verb.children[0].data == 'property_text':   
        prop_names, prop_obj_type = prop_or_verb.children[0].children
        prop_name = ':'
        for pname in prop_names.children:
            prop_name += f'{pname.value}_'
        print(prop_name[:-1])
        print(prop_obj_type.children[0].data)
    else:
        obj_type, text = prop_or_verb.children[0].children
        print(obj_type.data)
        print(text.children[0].value)
print()

rule = "dobj > :Ignore, dobj verb"
print('object', rule)
rule_parse = dsl.parse(rule)
tree = rule_parse.children[0].children[0] 
print(tree)
entity_tree = tree.children[0]
obj1_tree = tree.children[1]
obj2_tree = tree.children[2]
class_tree = tree.children[3]
prop_or_verbs = []
if len(tree.children) > 4:
    for i in range(4, len(tree.children)):
        prop_or_verbs.append(tree.children[i]) 
if entity_tree:
    print(entity_tree.children[0].data)
print(obj1_tree.children[0].data)
if obj2_tree:
    print(obj2_tree.children[0].data)
print(class_tree.children[0].value)
for prop_or_verb in prop_or_verbs:
    if prop_or_verb.children[0].data == 'property_text':   
        prop_names, prop_obj_type = prop_or_verb.children[0].children
        prop_name = ':'
        for pname in prop_names.children:
            prop_name += f'{pname.value}_'
        print(prop_name[:-1])
        print(prop_obj_type.children[0].data) 
    else:
        obj_type, text = prop_or_verb.children[0].children
        print(obj_type.data)
        print(text.children[0].value)   
print()
```
