# Test code to extract info from DSL entries

```python
def get_class_names(class_tree: Tree) -> list:
    class_names = []
    for i in range(0, len(class_tree.children)):
        class_names.append(f':{class_tree.children[i].children[0].children[0].value}')
    return class_names

def get_needed_text(text_tree: Tree) -> str:
    needed_text = ''
    for j in range(0, len(text_tree.children)):
        needed_text += text_tree.children[j].value + ' '
        if needed_text:
            needed_text = needed_text[:-1]
    return needed_text
```

```python
rule = ":Possession, :Continuation"
print(f'class_names rule: {rule}')
rule_parse = dsl.parse(rule)
class_tree = rule_parse.children[0].children[0]   # 1st level child = rule, 2nd = class_names
class_names = get_class_names(class_tree)
print()
```

```python
rules = [":Person . affiliation ArmyOfficer ArmedForce",
         "'army' > :Person, :Male",
         "'army' > :Person, :Male . affiliation ArmyOfficer ArmedForce"]
for rule in rules:
    print(f'noun rule: {rule}')
    rule_parse = dsl.parse(rule)
    tree = rule_parse.children[0].children[0]   # 1st level child = rule, 2nd = noun_rule
    increment = 0
    needed_text = ''
    if tree.children[0].data == 'text':        # May have text to check for
        increment = 1
        needed_text = get_needed_text(tree.children[0])
    noun_classes = get_class_names(tree.children[increment])
    affiliation_details = []
    if (needed_text and len(tree.children) == 3) or (not needed_text and len(tree.children) == 2):
        affiliation_details.append(tree.children[increment + 1].children[0].children[0].value)
        affiliation_details.append(tree.children[increment + 1].children[1].children[0].value)
    print(f'noun classes: {noun_classes}, needed text: {needed_text}, affiliation details: {affiliation_details}')
    print()
```

```python
rule = "('army' > :Person . affiliation ArmyOfficer ArmedForce) | ('navy' > :Person . affiliation NavyOfficer ArmedForce)"
print(f'complex rule: {rule}')
rule_parse = dsl.parse(rule)
tree = rule_parse.children[0].children[0]  # 1st level child = rule, 2nd = complex_rule
for i in range(0, len(tree.children)):     # Other children are different parentheses rules
    print(f'paren rule: {tree.children[i]}')
print()
```

```python
rules = ["'lack of judgment' > :Mistake", 
         "'back' & dobj > :ReturnRecoveryAndRelease ; :has_topic dobj",
         "dobj & pobj > :Storage ; :has_topic dobj ; :has_instrument pobj",
         "dobj > :Ignore, dobj verb",
         "Agent dobj > :CaringForDependents ; :has_affected_agent dobj"]
for rule in rules:
    print(f'verb rule: {rule}')
    rule_parse = dsl.parse(rule)
    tree = rule_parse.children[0].children[0]  # 1st level child = rule, 2nd = verb_rule
    needed_text = ''
    objects = []
    verb_classes = []
    property_details = []
    for i in range(0, len(tree.children)):
        child = tree.children[i]
        if child.data == 'text':        # May have text to check for
            needed_text = get_needed_text(child)
        elif child.data == 'obj_detail':
            obj_text = ''
            for j in range(0, len(child.children)):
                if child.children[j].data == 'entity_type':
                    obj_text = child.children[j].children[0].data + ' '
                else:
                    obj_text += child.children[j].children[0].data
            objects.append(obj_text)
        elif child.data == 'class_names':
            verb_classes = get_class_names(child)
        elif child.data == 'property_detail':
            property_details.append(f'; :{child.children[0].children[0].children[0].value} {child.children[1].children[0].data}')
        elif child.data == 'other_verb':
            verb_classes.append(f'dobj {child.children[0].value}')
        else:
            print('error')
    print(f'needed text: {needed_text}, verb classes: {str(verb_classes)}')
    print(f'objects: {str(objects)}, property details: {str(property_details)}')
    print()
```

```python
rule = "xcomp > :Attempt, xcompattack"
print(f'xcomp rule: {rule}')
rule_parse = dsl.parse(rule)
tree = rule_parse.children[0].children[0]
verb_classes = get_class_names(tree.children[0])
verb_classes.append(tree.children[1].value)
print(f'verb classes: {str(verb_classes)}')
```

```python
rules = ["subj > :EnvironmentAndConditionLineOfBusiness ; :has_topic :Judiciary ; :word_detail 'attorney'",
         "subj > :EnvironmentAndCondition ; :has_holder subj"]

for rule in rules:
    print(f'subject rule: {rule}')
    rule_parse = dsl.parse(rule)
    tree = rule_parse.children[0].children[0]
    verb_classes = get_class_names(tree.children[0])
    predicate_details = []
    for i in range(1, len(tree.children)):
        print(tree.children[i].data)
        if tree.children[i].data == 'topic':
            predicate_details.append(f':has_topic :{tree.children[i].children[0].children[0].children[0].value}')
        elif tree.children[i].data == 'word_detail':  
            predicate_details.append(f':word_detail "{tree.children[i].children[0].value}"')                         
        elif tree.children[i].data == 'holder':
            predicate_details.append(f':has_holder :{tree.children[i].children[0].value}')        
        else:
            print('error')
    print(f'verb classes: {str(verb_classes)}, predicate details: {str(predicate_details)}')
    print()
```
